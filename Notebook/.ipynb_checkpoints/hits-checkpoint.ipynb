### Anomaly replacement

```python
df_transformed = df_clean.copy()
```

```python
df_transformed.head()
```

```python
# проверим наличие аномалий в фиче с цифрами (visit_number)
```

```python
df_transformed['hit_number'].info()
```

```python
df_transformed.hit_number.describe().apply(lambda x: f"{x:0.3f}")
```

```python
plt.figure(figsize=(4, 2.5))
plt.boxplot(df_transformed['hit_number'])
plt.title("anomaly");
```

```python
df_transformed.hit_number.value_counts(dropna=False).head(50)
```

```python
def calculate_outliers(data) :
    q25 = data.quantile(0.25)
    q75 = data.quantile(0.75)
    iqr = q75 - q25
    boundaries = (q25 - 1.5 * iqr, q75 + 1.5 * iqr)
    return boundaries
```

```python
boundaries = calculate_outliers(df_transformed['hit_number'])
is_outlier = (df_transformed['hit_number'] < boundaries[0]) | (df_transformed['hit_number'] > boundaries[1])
print ('Количество аномалий в hit_number', is_outlier.sum())
```

```python
# видно, что у всех имеются аномалии, которые следует обработать
```

```python
print ("Нормальные значения колонки hit_number должны быть в таких границах:", boundaries)
```

```python
# выведем строки с нашими аномалиями
```

```python
df_transformed[is_outlier]
```

```python
# заменим аномальные значения на верхнюю границу, потому что все аномалии выше ее
```

```python
df_transformed.loc[is_outlier, 'hit_number'] = int(boundaries[1])
```

```python
plt.figure(figsize=(4, 2.5))
plt.boxplot(df_transformed['hit_number'])
plt.title("anomaly");
```

```python
# приобразуем hit_number в категориальную переменную
```

```python
df_transformed.hit_number = df_transformed.hit_number.astype('category')
df_transformed.hit_number.info()
```

# Feature engineering

### New features

```python
df_no_anamaly = df_transformed.copy()
```

```python
df_no_anamaly.info()
```

```python
df_no_anamaly.head(50)
```

```python
# создадим новые фичи на основе даты
```

```python
df_no_anamaly['month'] = df_no_anamaly.hit_date.dt.month
df_no_anamaly['dayofweek'] = df_no_anamaly.hit_date.dt.dayofweek
df_no_anamaly['year'] = df_no_anamaly.hit_date.dt.year
```

```python
df_no_anamaly.head()
```

```python
df_no_anamaly.info()
```

```python
df_no_anamaly['month'] = df_no_anamaly['month'].astype('category')
df_no_anamaly['dayofweek'] = df_no_anamaly['dayofweek'].astype('category')
df_no_anamaly['year'] = df_no_anamaly['year'].astype('category')
```

```python
df_no_anamaly.info()
```

```python
# постараемся достать информацию о марке машины и модели из hit_page_path
```

```python
df_no_anamaly.hit_page_path.head(10)
```

```python
manufacturer = df_no_anamaly.hit_page_path.apply(lambda x: x.split("/")[3] if len(x.split("/")) > 3 else None)
```

```python
manufacturer.value_counts(dropna=False).head(50)
```

```python
manufacturer.apply(lambda x: x.split("?")[0] if x is not None else None).value_counts(dropna=False).head(50)
```

```python
df_no_anamaly["manufacturer"] = manufacturer.apply(lambda x: x.split("?")[0] if x is not None else "other")
```

```python
model = df_no_anamaly.hit_page_path.apply(lambda x: x.split("/")[4] if len(x.split("/")) > 4 else None)
```

```python
model.value_counts(dropna=False).head(50)
```

```python
df_no_anamaly["model"] = df_no_anamaly.hit_page_path.apply(lambda x: x.split("/")[4] if len(x.split("/")) > 4 else "other")
```

```python
df_no_anamaly.head(50)
```

```python
df_no_anamaly.info()
```

```python
print_useful_rows_info(df_no_anamaly)
```

```python
# тперь обработаем изменим тип марки и производителя на категориальный
```

```python
df_no_anamaly['manufacturer'] = df_no_anamaly['manufacturer'].astype('category')
df_no_anamaly['model'] = df_no_anamaly['model'].astype('category')
```
