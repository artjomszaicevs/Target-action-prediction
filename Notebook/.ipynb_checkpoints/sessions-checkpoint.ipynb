### Anomaly replacement

```python
# проверим наличие аномалий в фиче с цифрами (visit_number)
```

```python
df_ses_transformed.visit_number.info()
```

```python
df_ses_transformed['visit_number'].nunique()
```

```python
df_ses_transformed['visit_number']
```

```python
plt.figure(figsize=(4, 2.5))
plt.boxplot(df_ses_transformed['visit_number'])
plt.title("anomaly");
```

```python
df_ses_transformed.visit_number.value_counts(dropna=False).head(50)
```

```python
def calculate_outliers(data) :
    q25 = data.quantile(0.25)
    q75 = data.quantile(0.75)
    iqr = q75 - q25
    boundaries = (q25 - 1.5 * iqr, q75 + 1.5 * iqr)
    return boundaries
```

```python
boundaries = calculate_outliers(df_ses_transformed['visit_number'])
# print (boundaries)
is_outlier = (df_ses_transformed['visit_number'] < boundaries[0]) | (df_ses_transformed['visit_number'] > boundaries[1])
print ('Количество аномалий в visit_number', is_outlier.sum())
```

```python
# видно, что у всех имеются аномалии, которые следует обработать
```

```python
print ("Нормальные значения колонки visit_number должны быть в таких границах:", boundaries)
```

```python
# выведем строки с нашими аномалиями
```

```python
df_ses_transformed[is_outlier]
```

```python
df_ses_transformed.visit_number.describe().apply(lambda x: f"{x:0.3f}")
```

```python
# заменим аномальные значения на верхнюю границу, потому что все аномалии выше ее
```

```python
df_ses_transformed.loc[is_outlier, 'visit_number'] = int(boundaries[1])
```

```python
plt.figure(figsize=(4, 2.5))
plt.boxplot(df_ses_transformed['visit_number'])
plt.title("anomaly");
```

```python
is_outlier = (df_ses_transformed['visit_number'] < boundaries[0]) | (df_ses_transformed['visit_number'] > boundaries[1])
print ('Количество аномалий в visit_number', is_outlier.sum())
```

# Feature engineering

### New features

```python
df_no_anamaly.info()
```

```python
df_no_anamaly.head(50)
```

```python
# начну с того, что создам две новые переменные на основе device_screen_resolution
```

```python
df_no_anamaly['device_screen_width'] = df_no_anamaly['device_screen_resolution'].apply(lambda x: x.split('x')[0])
```

```python
df_no_anamaly['device_screen_height'] = df_no_anamaly['device_screen_resolution'].apply(lambda x: x.split('x')[1] if 'x' in x and len(x.split('x')) == 2 else None)
```

```python
df_no_anamaly.head(50)
```

```python
missing_values = ((df_no_anamaly.isna().sum() / len(df_no_anamaly)) * 100).sort_values()
print ('Процент пропущенных значений:')
missing_values
```

```python
df_no_anamaly[df_no_anamaly.device_screen_height.isna()]
```

```python
# создадим новые фичи на основе даты
```

```python
df_no_anamaly['month'] = df_no_anamaly.date_and_time.dt.month
df_no_anamaly['dayofweek'] = df_no_anamaly.date_and_time.dt.dayofweek
df_no_anamaly['year'] = df_no_anamaly.date_and_time.dt.year
df_no_anamaly['hour'] = df_no_anamaly.date_and_time.dt.hour
```
